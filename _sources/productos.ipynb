# Catalogo de productos

```python
import matplotlib.pyplot as plt
data = [
    ["Nombre", "Descripción", "Precio"],
    ["Producto 1", "Descripción 1", 19.99],
    ["Producto 2", "Descripción 2", 29.99],
    ["Producto 3", "Descripción 3", 14.99],
    ["Producto 4", "Descripción 4", 24.99],
    ["Producto 5", "Descripción 5", 9.99]
]
fig, ax = plt.subplots()
ax.axis('off')  # Oculta los ejes

tabla = ax.table(cellText=data, loc='center', cellLoc='center', colWidths=[0.2, 0.4, 0.2])
tabla.auto_set_font_size(False)
tabla.set_fontsize(12)
tabla.scale(1.5, 1.5)  # Ajusta el tamaño de la tabla
plt.show()


```
